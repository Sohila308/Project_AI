```python
# Splitting Data into (Test & Training)
X_train,X_test,Y_train,Y_test =train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
print(Y.shape,Y_train.shape,Y_test.shape)
```

```python
# Using Machine Learning Algorithm (first algorithm "Logistic Regression")
# Train the model
model = LogisticRegression()
model.fit(X_train, Y_train)

# === Evaluation on Training Data ===
train_preds = model.predict(X_train)
train_accuracy = accuracy_score(Y_train, train_preds)
train_cm = confusion_matrix(Y_train, train_preds)

print("Accuracy on Training Data =", train_accuracy)
print("Confusion Matrix on Training Data:")
print(train_cm)

# Plot confusion matrix for training data 
plt.figure(figsize=(6, 4))
sns.heatmap(train_cm, annot=True, fmt='d', cmap='YlGn', xticklabels=[0, 1], yticklabels=[0, 1])
plt.title("Confusion Matrix - Training Data")
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.show()

# === Evaluation on Test Data ===
test_preds = model.predict(X_test)
test_accuracy = accuracy_score(Y_test, test_preds)
test_cm = confusion_matrix(Y_test, test_preds)

print("Accuracy on Test Data =", test_accuracy)
print("Confusion Matrix on Test Data:")
print(test_cm)

# Plot confusion matrix for test data 
plt.figure(figsize=(6, 4))
sns.heatmap(test_cm, annot=True, fmt='d', cmap='Oranges', xticklabels=[0, 1], yticklabels=[0, 1])
plt.title("Confusion Matrix - Test Data")
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.show()

# Classification Report on Test Data
print("Classification Report on Test Data:")
print(classification_report(Y_test, test_preds))
```

```python
# Using Machine Learning Algorithm (Second algorithm "RandomForest")

# Create and train the model
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, Y_train)

# === Evaluation on Training Data ===
train_preds = model.predict(X_train)
train_accuracy = accuracy_score(Y_train, train_preds)
train_cm = confusion_matrix(Y_train, train_preds)

print("Accuracy on Training Data =", train_accuracy)
print("Confusion Matrix on Training Data:")
print(train_cm)

# Plot confusion matrix for training data 
plt.figure(figsize=(6, 4))
sns.heatmap(train_cm, annot=True, fmt='d', cmap='YlGn', xticklabels=[0, 1], yticklabels=[0, 1])
plt.title("Confusion Matrix - Training Data")
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.show()

# === Evaluation on Test Data ===
test_preds = model.predict(X_test)
test_accuracy = accuracy_score(Y_test, test_preds)
test_cm = confusion_matrix(Y_test, test_preds)

print("Accuracy on Test Data =", test_accuracy)
print("Confusion Matrix on Test Data:")
print(test_cm)

# Plot confusion matrix for test data 
plt.figure(figsize=(6, 4))
sns.heatmap(test_cm, annot=True, fmt='d', cmap='Oranges', xticklabels=[0, 1], yticklabels=[0, 1])
plt.title("Confusion Matrix - Test Data")
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.show()

# Classification Report
print("Classification Report on Test Data:")
print(classification_report(Y_test, test_preds))
```

```python
# Using Machine Learning Algorithm (Third algorithm "SVC")

# --------- 1. Scaling the data ---------
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# --------- 2. Create and Train the SVM model ---------
model = SVC(kernel='linear', C=1)
model.fit(X_train_scaled, Y_train)

# --------- 3. Evaluation on Training Data ---------
train_preds = model.predict(X_train_scaled)
train_accuracy = accuracy_score(Y_train, train_preds)
print("Accuracy on Training Data =", train_accuracy)

train_cm = confusion_matrix(Y_train, train_preds)
print("Confusion Matrix on Training Data:")
print(train_cm)

# Plot confusion matrix for training data 
plt.figure(figsize=(6, 4))
sns.heatmap(train_cm, annot=True, fmt='d', cmap='YlGn', xticklabels=[0, 1], yticklabels=[0, 1])
plt.title("Confusion Matrix - Training Data")
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.show()

# --------- 4. Evaluation on Test Data ---------
test_preds = model.predict(X_test_scaled)
test_accuracy = accuracy_score(Y_test, test_preds)
print("Accuracy on Test Data =", test_accuracy)

test_cm = confusion_matrix(Y_test, test_preds)
print("Confusion Matrix on Test Data:")
print(test_cm)

# Plot confusion matrix for test data 
plt.figure(figsize=(6, 4))
sns.heatmap(test_cm, annot=True, fmt='d', cmap='Oranges', xticklabels=[0, 1], yticklabels=[0, 1])
plt.title("Confusion Matrix - Test Data")
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.show()

# --------- 5. Detailed Classification Report ---------
print("Classification Report on Test Data:")
print(classification_report(Y_test, test_preds))
```
